```python
import seaborn as sns
import numpy as np
import json
from pprint import pprint
import matplotlib.pyplot as plt
```

```python
full_pretrain_perf = []

# open the file with the performances over 100 epochs using the pretrained model
with open("../50_episodes_1_9/run_1/full_pretrain_performances.json", "r") as full_pretrain_perf_file:
    full_pretrain_perf.extend(json.load(full_pretrain_perf_file))
    
# open the file with the performances over 100 epochs using the pretrained model
with open("../50_episodes_1_9/run_2/full_pretrain_performances.json", "r") as full_pretrain_perf_file:
    full_pretrain_perf.extend(json.load(full_pretrain_perf_file))

# open the file with the performances over 100 epochs using the pretrained model
with open("../50_episodes_1_9/run_3/full_pretrain_performances.json", "r") as full_pretrain_perf_file:
    full_pretrain_perf.extend(json.load(full_pretrain_perf_file))



full_scratch_perf = []
    
# open the file with the performances over 100 epochs using the scratch model
with open("../50_episodes_1_9/run_1/full_scratch_performances.json", "r") as full_scratch_perf_file:
    full_scratch_perf.extend(json.load(full_scratch_perf_file))
    
with open("../50_episodes_1_9/run_2/full_scratch_performances.json", "r") as full_scratch_perf_file:
    full_scratch_perf.extend(json.load(full_scratch_perf_file))
    
with open("../50_episodes_1_9/run_3/full_scratch_performances.json", "r") as full_scratch_perf_file:
    full_scratch_perf.extend(json.load(full_scratch_perf_file))
    

    
print(len(full_pretrain_perf))
print(len(full_scratch_perf))
```

```python
# they have to be in columns, not in rows

full_pretrain_mean = []
full_scratch_mean = []

for i in range(160):
    full_pretrain_mean.append([full_pretrain_perf[i][str(j)] for j in range(50)])
    full_scratch_mean.append([full_scratch_perf[i][str(j)] for j in range(50)])
```

```python
plt.figure(figsize=(15, 9))
sns.set(font_scale=3)
sns.set_style("whitegrid")


data = np.asarray(full_pretrain_mean)
ax = sns.tsplot(data=data,  ci=[90], color='blue', marker='d', markersize=15, markevery=5, linewidth=3.0, legend=True, condition="transfer learning, no warm-start")

data = np.asarray(full_scratch_mean)
ax = sns.tsplot(data=data,  ci=[90], color='red', marker='X', markersize=15, markevery=5, linewidth=3.0, legend=True, condition="no transfer learning, warm-start")



ax.set_xlabel('Number of Epochs', weight='bold', size=35)
ax.set_ylabel('Success Rate', weight='bold', size=35)
sns.plt.title('Learning curve over training data set', weight='bold', size=35)

plt.xlim((0, 50))
plt.ylim((0, 0.42))
plt.setp(ax.get_legend().get_texts(), fontsize="35")
plt.savefig('learning_curve_training_data_set_200_reps_color.png', dpi=200, bbox_inches="tight", pad_inches=0)
plt.show()
```
