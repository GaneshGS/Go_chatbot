```python
import seaborn as sns
import numpy as np
import json
from pprint import pprint
import matplotlib.pyplot as plt
```

## Training Performance Over Epochs

```python
# open the file with the performances over 100 epochs using the pretrained model
with open("../100_episodes_1_9/run_1/full_pretrain_performances.json", "r") as full_pretrain_perf_file:
    full_pretrain_perf = json.load(full_pretrain_perf_file)
    
# open the file with the performances over 100 epochs using the scratch model
with open("../100_episodes_1_9/run_1/full_scratch_performances.json", "r") as full_scratch_perf_file:
    full_scratch_perf = json.load(full_scratch_perf_file)
```

```python
# they have to be in columns, not in rows

full_pretrain_perf_mean = []
full_scratch_perf_mean = []

for i in range(20):
    full_pretrain_perf_mean.append([full_pretrain_perf[i][str(j)] for j in range(100)])
    full_scratch_perf_mean.append([full_scratch_perf[i][str(j)] for j in range(100)])
```

```python
plt.figure(figsize=(7, 5))
data = np.asarray(full_pretrain_perf_mean)
ax = sns.tsplot(data=data,  ci=[90], color='blue', linewidth=3.0)
data = np.asarray(full_scratch_perf_mean)
ax = sns.tsplot(data=data,  ci=[90], color='red', linewidth=3.0)
plt.show()
```

## Testing performance over epochs

```python
# open the file with the performances over 100 epochs using the pretrained model
with open("../100_episodes_1_9/run_1/full_pretrain_test_performances.json", "r") as full_pretrain_test_perf_file:
    full_pretrain_test_perf = json.load(full_pretrain_test_perf_file)
    
# open the file with the performances over 100 epochs using the scratch model
with open("../100_episodes_1_9/run_1/full_scratch_test_performances.json", "r") as full_scratch_test_perf_file:
    full_scratch_test_perf = json.load(full_scratch_test_perf_file)
```

```python
# they have to be in columns, not in rows

full_pretrain_test_perf_mean = []
full_scratch_test_perf_mean = []

for i in range(20):
    full_pretrain_test_perf_mean.append([full_pretrain_test_perf[i][str(j)] for j in range(100)])
    full_scratch_test_perf_mean.append([full_scratch_test_perf[i][str(j)] for j in range(100)])
```

```python
plt.figure(figsize=(10, 7))
data = np.asarray(full_pretrain_test_perf_mean)
ax = sns.tsplot(data=data,  ci=[90], color='blue', linewidth=3.0)


data = np.asarray(full_scratch_test_perf_mean)
ax = sns.tsplot(data=data,  ci=[90], color='red', linewidth=3.0)


sns.plt.legend(labels=["transfer learning", "no transfer learning"])

ax.set_xlabel('Number of epochs', weight='bold', size=15)
ax.set_ylabel('Success Rate', weight='bold', size=15)
sns.plt.title('Testing Performance for 5 user goals', weight='bold', size=15)

plt.savefig("learning_curve_20_episodes_20_repetitions_95_confidence.png")
```
